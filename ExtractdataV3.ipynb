```python
import pandas as pd
import re
from itables import show

df1 = pd.read_csv("binary_5plus_high_row_count_part1.csv")
df2 = pd.read_csv("binary_5plus_high_row_count_part2.csv")
# Combine with df2 below df1
df = pd.concat([df1, df2], axis=0, ignore_index=True)

```

```python
# Remove columns where ALL values are NaN/empty
df = df.dropna(axis=1, how='all')
```

```python
print(df.shape)
print(df.iloc[1])
show(df.iloc[1])
```

```python
print(df.columns.tolist())
```

```python
print(df["phase"].unique())
```

```python
df_v1 = df[df['phase'].isin(['Gas','Liquid'])]
print(f"Shape after dropping: {df_v1.shape}")
```

```python
print(df_v1["property"].unique())
```

```python
# Drop rows where Temperature,K is NaN
df_v2 = df_v1.dropna(subset=['Temperature, K'], how='all')

print(f"Shape after dropping: {df_v2.shape}")
```

```python
print(df_v2["property"].unique())
```

```python
# Find elements in df but not in df_v1
diff = set(df_v1["property"].unique()) - set(df_v2["property"].unique())
print("Elements removed:", sorted(diff))
```

```python
# Drop rows where ALL specified columns about Pressure (P) are NaN
df_v3 = df_v2.dropna(subset=['Pressure, kPa','Pressure, kPa_1', 'Pressure, kPa_2'], how='all')

print(f"Shape after dropping: {df_v3.shape}")
```

```python
print(df_v3["property"].unique())
```

```python
# Find elements in df_v1 but not in df_v2
diff = set(df_v2["property"].unique()) - set(df_v3["property"].unique())
print("Elements removed:", sorted(diff))
```

```python
# Remove columns where ALL values are NaN/empty
df_v4 = df_v3.dropna(axis=1, how='all')

print(f"Shape after dropping: {df_v4.shape}")
print(df_v4.columns.tolist())
```

```python
print(df_v4["property"].unique())
```

```python
# Drop rows where ALL specified columns about mole fractions are NaN
df_v4.loc[df_v4['property'] == 'mole fraction', 'Mole fraction'] = df_v4['value']
df_v5 = df_v4.dropna(subset=['Mole fraction'], how='all')

print(f"Shape after dropping: {df_v5.shape}")
```

```python
print(df_v5["property"].unique())
# Find elements in df_v1 but not in df_v2
diff = set(df_v4["property"].unique()) - set(df_v5["property"].unique())
print("Elements removed:", sorted(diff))
```
