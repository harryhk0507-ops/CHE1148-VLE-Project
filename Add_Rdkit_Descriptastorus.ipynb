```python
import pandas as pd
import re
import numpy as np
import torch
import random
from itables import show
import pubchempy as pcp
```

```python
df = pd.read_csv("Cleaned_VLE_Data.csv")
```

```python
# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)
```

```python
print(df.shape)
df.head()
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)

def get_smiles(name):
    try:
        # 1. Call the API only once
        compounds = pcp.get_compounds(name, 'name')
        
        # 2. Check if results exist
        if compounds:
            return compounds[0].isomeric_smiles
            
    except Exception as e:
        # 3. Handle network errors (timeouts, connection issues)
        print(f"Error fetching {name}: {e}")
        
    # Return None (or "Unknown") if no compound found or error occurred
    return None 

# Apply the optimized function
df['Smiles 1'] = df['Component 1'].apply(get_smiles)
```
